```python
import numpy
import scipy
from scipy import ndimage, misc
```

```python
# use sample image of racoon face from misc package
# convert to grayscale
racoon = misc.face(gray=True)

# convert to float for pyplot
racoon=racoon.astype('float32')
```

```python
# horizontal gradient
dx = ndimage.sobel(racoon, axis=0, mode='constant')

# vertical gradient
dy = ndimage.sobel(racoon, axis=1, mode='constant')

# gradient magnitude
mag = numpy.hypot(dx, dy)
```

```python
# import and setup plots
import matplotlib.pyplot as plt
plt.figure()
plt.gray()
```

```python
# plot original image
plt.subplot(111)
plt.imshow(racoon, cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```

```python
# plot horizontal and vertical gradients
plt.subplot(121)
plt.imshow(dx, cmap=plt.cm.gray)
plt.axis('off')
plt.subplot(122)
plt.imshow(dy, cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```

```python
# plot final image
plt.subplot(111)
plt.imshow(mag, cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```
