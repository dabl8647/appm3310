```python
import numpy
import scipy
from scipy import ndimage, misc
from scipy.ndimage import gaussian_filter
```

```python
# Note make sure to convert to grayscale

# use sample image of racoon face from misc package
image = misc.face(gray=True)

# use sample from desktop (change to an appropriate path)
#image = misc.imread("/Users/bloomdt/Desktop/engineering_building.png", flatten=True)

# convert to float for pyplot
image = image.astype('float32')

# use gaussian blur
blur_image = gaussian_filter(image, sigma=1.5)
```

```python
# horizontal gradient - estimating the gradient in the x-direction (columns), so dx detects vertical lines.
dx = ndimage.sobel(image, axis=0, mode='constant')

# vertical gradient - estimating the gradient in the y-direction (rows), so dy detects horizontal lines
dy = ndimage.sobel(image, axis=1, mode='constant')

# gradient magnitude
mag = numpy.hypot(dx, dy)
```

```python
# horizontal gradient on blur- estimating the gradient in the x-direction (columns), so dx detects vertical lines.
blur_dx = ndimage.sobel(blur_image, axis=0, mode='constant')

# vertical gradient on blur - estimating the gradient in the y-direction (rows), so dy detects horizontal lines
blur_dy = ndimage.sobel(blur_image, axis=1, mode='constant')

# gradient magnitude on blur
blur_mag = numpy.hypot(blur_dx, blur_dy)
```

```python
# import and setup plots

import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,15))
plt.gray()

# plot original and blur images

plt.subplot(121)
plt.imshow(image, cmap=plt.cm.gray)
plt.title('original')
plt.axis('off')

plt.subplot(122)
plt.imshow(blur_image, cmap=plt.cm.gray)
plt.title('blur')
plt.axis('off')

plt.show()
```

```python
# plot vertical gradients

plt.figure(figsize=(15,15))
plt.gray()

plt.subplot(121)
plt.imshow(dx, cmap=plt.cm.gray)
plt.title('vertical lines')
plt.axis('off')

plt.subplot(122)
plt.imshow(blur_dx, cmap=plt.cm.gray)
plt.title('vertical lines blur')
plt.axis('off')

plt.show()

# plot horizontal gradients

plt.figure(figsize=(15,15))
plt.gray()

plt.subplot(121)
plt.imshow(dy, cmap=plt.cm.gray)
plt.title('horizontal lines')
plt.axis('off')

plt.subplot(122)
plt.imshow(blur_dy, cmap=plt.cm.gray)
plt.title('horizontal lines blur')
plt.axis('off')

plt.show()
```

```python
# plot final images

plt.figure(figsize=(15,15))
plt.gray()

plt.subplot(121)
plt.imshow(mag, cmap=plt.cm.gray)
plt.title('edge detection')
plt.axis('off')

plt.subplot(122)
plt.imshow(blur_mag, cmap=plt.cm.gray)
plt.title('edge detection blur')
plt.axis('off')

plt.show()
```
